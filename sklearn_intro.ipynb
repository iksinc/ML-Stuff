### Intro to SKLearn
- Scikit-learn is an open source machine learning library that supports supervised and unsupervised learning. 
It also provides various tools for model fitting, data preprocessing, model selection, model evaluation,
and many other utilities.
- It provides numerous built-in machine learning algorithms and models, called ${estimators}$. You can fit these estimators to some data using its fit method. Fitting means building the model.
- SKLearn also comes with some well-known data sets for experimentation. You cal also generate data for experimentation.
- SKLearn also offers preprocessors and transformers to massage input data before it is used for model building
- It also includes tools to perform training via train-test splits as well as tools to evaluate the fitted models
- Many machine learning models need parameters set by the user for optimal performance. SKLearn offers ways to do parameter search in a systematic way.
- Finally, you can assemble your workflow  via the $pipeline$ object of SKLearn

```python
# An example of SKLearn usage
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(noise=0.30, random_state=3, n_samples=100)
plt.scatter(X[:, 0], X[:, 1], c=y, marker="o", s=25,)
plt.title('Plot of Make_Moons Data')
plt.show()
```

```python
from sklearn.neighbors import NearestCentroid
clf = NearestCentroid()
clf.fit(X, y)
print(clf.predict([[-0.1, 0.5]]))
```

```python
#Visualize the fitted model
# Plot the decision boundary
from matplotlib.colors import ListedColormap
import numpy as np
cmap_light = ListedColormap(['#FFAAAA','#AAFFAA', '#AAAAFF'])
x_min,x_max = X[:,0].min()-1, X[:,0].max()+1
y_min,y_max = X[:,0].min()-1, X[:,0].max()+1
xx,yy = np.meshgrid(np.arange(x_min,x_max, 0.02),
np.arange(y_min, y_max, 0.02))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx,yy,Z,shading='nearest', vmin=Z.min()-1, vmax=Z.max()+1)
plt.scatter(X[:,0], X[:,1], c=y, s=25)
plt.xlabel("first feature")
plt.ylabel("second feature")
plt.show()
```
